```python
import pandas as pd

games_tweets_df = pd.read_csv('Datasets\\tweets_games_final.csv', index_col=0)
games_tweets_df
```

## Cleaning

```python
# Dropping NA
gt_df = games_tweets_df.dropna()
gt_df.head()
```

Doing so dropped the majority of the columns. Looks like most of the tweets in the dataset are not closely assocaited with a following game day. There may be some parameter adjustments needed to be made to account for this better.

### Cleaning parenthesis out of string numbers and mapping L = 0, W = 1

```python
# Cleaning parentheses out of helpful quantitative values

gt_df['Leading Scorer Points'] = gt_df['Leading Scorer Points'].str.replace(r'[()]', '')
gt_df['Leading Scorer Points'] = gt_df['Leading Scorer Points'].astype(int)
gt_df['Leading Scorer Points.1'] = gt_df['Leading Scorer Points.1'].str.replace(r'[()]', '')
gt_df['Leading Scorer Points.1'] = gt_df['Leading Scorer Points.1'].astype(int)
gt_df.head()
```

```python
gt_df.dtypes
```

```python
# Mapping Label column to 0 and 1

gt_df['following_result'].map({'L': 0, 'W': 1})
gt_df.head()
```

```python
gt_df['following_result'].value_counts()
```

#### Splitting out Feature and Label Columns

```python
feature_df = gt_df[['verified', 'retweet_count', 'like_count', 'Opposing Team', 'Leading Scorer-Winner', 'Leading Scorer Points', 'Leading Scorer Loss', 'Leading Scorer Points.1', 'tweet_sentiment']]
feature_df
```

```python
label_df = gt_df['following_result']
label_df
```

### Further Feature Engineering: One-hot encoding on categorical variables

```python
from sklearn.preprocessing import OneHotEncoder

feature_df['Opposing Team'] = feature_df['Opposing Team'].astype('category')
feature_df['Leading Scorer-Winner'] = feature_df['Leading Scorer-Winner'].astype('category')
feature_df['Leading Scorer Loss'] = feature_df['Leading Scorer Loss'].astype('category')
feature_df = pd.get_dummies(feature_df)
feature_df
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
to_scale = ['retweet_count', 'like_count', 'Leading Scorer Points', 'Leading Scorer Points.1', 'tweet_sentiment']
feature_df[to_scale] = scaler.fit_transform(feature_df[to_scale])
feature_df.head()
```

## Model Selection and Training

Since the data is rather sparse due to the one-hot encoding, going to try some L1 Regularization models as well as some trees for power.

Cross-validating with k=5 throughout.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

lr_l1 = LogisticRegression(C=1, penalty='l1', solver='liblinear')
svm_linear = SVC(kernel="linear")
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(random_state=23) # some random seed for reproducibility
grad_boost = GradientBoostingClassifier()

all_models = {"lr":lr_l1, 
              "svm":svm_linear,
              "decision_tree":dt,
              "random_forest":rf,
              "grad_boost":grad_boost
              }

print (f"We are working with classifiers {all_models.keys()}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

features = feature_df.to_numpy()
labels = label_df.to_numpy()
_x, x_test, _y, y_test = train_test_split(features, labels, test_size=0.10, random_state=42)

k = 5 # 5-fold

best_model_name = ""
best_model_valid_accuracy = 0
best_model = None

for model_name in all_models.keys():
    model = all_models[model_name]
    cv_scores = cross_val_score(model,_x,_y.flatten(), cv=k, n_jobs=4)
    average_cv_score = cv_scores.mean()
    print (f"Mean cross validation accuracy for model {model_name} = {average_cv_score}")

    if average_cv_score > best_model_valid_accuracy :
        best_model_name = model_name
        best_model_valid_accuracy  = average_cv_score
        best_model = model

print (f"Best model is {best_model_name} with {k}-fold accuracy of {best_model_valid_accuracy}")
```

The model has a suspiciously high accuracy. This lends into potential overfitting.

### Testing accuracy and Classification Metrics

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# Fitting best model again

best_model.fit(_x,_y.flatten())

y_pred_test = best_model.predict(x_test)
test_accuracy = accuracy_score(y_pred_test, y_test.flatten())

print (f"Test accuracy for model {test_accuracy}")
print (classification_report(y_test.flatten(),y_pred_test))
```

While the metrics do not immediately clue into overfitting and the resultsare impressive, this does not mean that there is something strange happening with the predictions - perhaps the model is learning to use some overlooked information or looking ahead to predict the outcome of the game. To try and solve for this, going to perform some feature selection/ablation testing to identify the most important features.
