Sentiment analysis and additional dataset manipulation of tweets regarding NBA games.

```python
import pandas as pd
import datetime as datetime
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
import numpy as np
import matplotlib.pyplot as plt
import tqdm
import math
```

Done importing.

```python
tweets_df = pd.read_csv('Datasets\mavs_games_tweets_merged.csv')
tweets_df.head()
```

```python
tw_text = tweets_df['text']
tw_text.head()
tw_text
```

Creating a sentiment analyzer and appending sentiment to the main dataset

```python
sia = SentimentIntensityAnalyzer()

score = []
for tweet in tqdm.tqdm(tw_text):
    tweet = str(tweet)
    sentiment = sia.polarity_scores(tweet)
    compoundScore = score.append(sentiment['compound'])

tweets_df['tweet_sentiment'] = score
tweets_df
```

```python
tweets_sent = pd.read_csv('Datasets\\tweets_games_sentiment.csv')

#additional cleaning: reversing and then sorting by datetime
tweets_sent = tweets_sent.iloc[::-1]

tweets_sent = tweets_sent.sort_values(by='Datetime')
tweets_sent = tweets_sent.reset_index(drop=True)

tweets_sent
```

This next bit of code will create a new column with game results of the next game if the tweet originally has no game associated with it. Could be made a bit more efficient.

```python
game_wl = tweets_sent['Win or Loss']
game_wl = game_wl.replace({np.nan: None})

record = pd.read_csv('Datasets\\Mavs_game_Data.csv')
record = record['Win or Loss']
iter_record = iter(record)
nextval = next(iter_record)

final_wl = []

while game_wl.isna().any():
    #Searching for the first non null index and filling the list with the following W/L value, then dropping the nulls
    first_index = game_wl.first_valid_index()
    if first_index != 0:
        print(first_index)
        first_result = game_wl[first_index]
        final_wl += first_index * [first_result]
        game_wl.drop(index=game_wl.index[:first_index], inplace=True)
        game_wl = game_wl.reset_index(drop=True)

    #Finding the non-null W/L values and adding it to the list. If null, then breaks
    for i, value in game_wl.items():
        if value is None:
            break
        final_wl.append(value)
        game_wl.drop(index=game_wl.index[i], inplace=True)
        game_wl = game_wl.reset_index(drop=True)
```

```python
#Adding in the rest of the Null's and adding the values as a column
final_wl += [None] * 24
tweets_sent['following_result'] = final_wl
```
